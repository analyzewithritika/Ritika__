```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
academics = pd.read_excel(r"C:\Users\rawat\Downloads\academics.xlsx")
academics.head(100)
```

```python
# Plotting
plt.figure(figsize=(10, 6))
plt.bar(academics['Low Performers (%)'],academics['Average Grade (%)'],academics['High Achievers (%)'],color=['red', 'yellow', 'green'])


# Adding titles and labels
plt.title('Total Students vs Achievers Categories')
plt.xlabel('Total Students')
plt.ylabel('Categories')

# Display the plot
plt.show()
```

```python
# Plotting
plt.figure(figsize=(10, 6))
plt.bar(academics['Extracurricular Participation (%)'],academics['Sports Participation (%)'],academics['Community Service Participation (%)'],color=['red', 'yellow', 'green'])


# Adding titles and labels
plt.title('Total Students vs Extracurricular Activities')
plt.xlabel('Total Students')
plt.ylabel('Categories')

# Display the plot
plt.show()
```
